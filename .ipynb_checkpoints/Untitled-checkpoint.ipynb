```python
df = pd.read_excel('resultados.xlsx', engine='openpyxl', skiprows=5)
df.head()
```

```python
df.columns = ['Concurso', 'Data', 'Número 1', 'Número 2', 'Número 3', 'Número 4', 'Número 5', 'Número 6']
```

```python
df.head()
```

```python
df.head(20)  # Exibe as 20 primeiras linhas para inspeção
```

```python
for col in ['Número 1', 'Número 2', 'Número 3', 'Número 4', 'Número 5', 'Número 6']:
    df[col] = pd.to_numeric(df[col], errors='coerce')  # Garante que são números
df = df[(df[['Número 1', 'Número 2', 'Número 3', 'Número 4', 'Número 5', 'Número 6']] >= 1).all(axis=1)]
df = df[(df[['Número 1', 'Número 2', 'Número 3', 'Número 4', 'Número 5', 'Número 6']] <= 60).all(axis=1)]
```
